```python
import pandas as pd
from openpyxl import load_workbook
from openpyxl.styles import PatternFill
df2022 = pd.read_excel('2022_年間売上表.xlsx', sheet_name='Sheet1')
df2023 = pd.read_excel('2023_年間売上表.xlsx', sheet_name='Sheet1')
df = pd.concat([df2022, df2023])
df_sum = df.groupby(['商品', '売上年'], as_index=False)['金額（千円）'].sum()
df_sum.to_excel('売上集計表.xlsx', index=False)
wb = load_workbook('売上集計表.xlsx')
ws = wb.active
fill = PatternFill(start_color='F2F2F2',
                   end_color='F2F2F2',
                   fill_type='solid')
for cell in ws[1]:
    cell.fill = fill
wb.save('売上集計表.xlsx')
```
